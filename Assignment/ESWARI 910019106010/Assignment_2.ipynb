# 9. Scale the independent variable

```python
def normalize(arr, t_min, t_max):
    norm_arr = []
    diff = t_max - t_min
    diff_arr = max(arr) - min(arr)
    for i in arr:
        temp = (((i - min(arr))*diff)/diff_arr) + t_min
        norm_arr.append(temp)
    return norm_arr
range_to_normalize= (0,1)
normalized_y = normalize(y, range_to_normalize[0], 
  range_to_normalize[1])
```

```python
# the independent variable (Exited) will be normalized between 0 to 1
normalized_y
```

```python
# we could see that the dependent variable is between 0 & 1
```
